```python
import numpy as np
import h5py
import math
import time,sys,os
import datetime
import matplotlib.pyplot as plt
from astropy.cosmology import Planck15, FlatLambdaCDM, z_at_value
import astropy.units as u
import lightcone as lc
import importlib
```

```python
snap_path = '/cosma7/data/dp004/dc-boot5/Ordered_Snapshots/Npart_2048_Box_3000-Fiducial/'
```

```python
def read_header(snap):
    fname = 'snapdir_{0:02d}/ordered_snapshot.snap_{0:03d}.00.hdf5'.format(snap)
    in_file = snap_path+fname
    with h5py.File(in_file, 'r') as f:
        h = f['Header']
        z = h.attrs['Redshift']
        print('Snapshot {0:02d}, redshift = {1:02.2f}'.format(snap,z))
```

```python
for snap in range(42,64):
    read_header(snap)
```

```python
# define particle datatype
vect = np.dtype([('x', np.float32),('y', np.float32),('z', np.float32)])
part = np.dtype([('pos', vect),('vel', vect),('ID', np.ulonglong)])
```

```python
def read_parts(snap):
    fname = 'snapdir_{0:02d}/ordered_snapshot.snap_{0:03d}.{1:02d}.hdf5'.format(snap, sub_snap)
    in_file = snap_path+fname
    isnap = snap - first_snap
    global pts
    with h5py.File(in_file, 'r') as f:
        h = f['Header']
        z = h.attrs['Redshift']
        p = f['Parts']
        pts[isnap,::] = p[:nparts]
        print('Snapshot {0:02d}, redshift = {1:02.2f}'.format(snap,z))
```

```python
nparts = 1000
nsnaps = 20
first_snap = 44
sub_snap = 0
pts = np.empty((nsnaps, nparts),dtype = part)
```

```python
for snap in range(first_snap, first_snap + nsnaps):
    read_parts(snap)
```
